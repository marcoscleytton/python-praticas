```python
import pandas as pd 
```

```python
# Lendo cada aba
df_pedidos = pd.read_excel('Superloja.xls', sheet_name='Pedidos')
df_devolucoes = pd.read_excel('Superloja.xls', sheet_name='Devoluções')
df_pessoas = pd.read_excel('Superloja.xls', sheet_name='Pessoas')

# Juntando pedidos com devoluções (ID do pedido)
df_merge = pd.merge(df_pedidos, df_devolucoes, on='ID do pedido', how='left')

# Juntando com pessoas (Região)
df_final = pd.merge(df_merge, df_pessoas, on='Região', how='left')
```

```python
vitor_resultado = (
    df_final[df_final["Gerente regional"].str.strip() == "Vitór Rodrigues"]
    .groupby(["Categoria", "Região"])
    .agg(
        Quantidade=("Devolvido", "count"),
        Taxa_devolucao=("Devolvido", "mean")
    )
    .reset_index()
)

# Produto mais devolvido
produto_top = (
    df_final[(df_final["Gerente regional"].str.strip() == "Vitór Rodrigues") & (df_final["Devolvido"] == 1)]
    .groupby(["Categoria", "Nome do produto"])["Devolvido"]
    .count()
    .reset_index()
)
produto_top = produto_top.loc[produto_top.groupby("Categoria")["Devolvido"].idxmax()]

# Junta e ordena pela quantidade (maior primeiro)
vitor_resultado = pd.merge(
    vitor_resultado,
    produto_top[["Categoria", "Nome do produto"]],
    on="Categoria",
    how="left"
).rename(columns={"Nome do produto": "Produto mais devolvido"})

vitor_resultado = vitor_resultado.sort_values("Quantidade", ascending=False).reset_index(drop=True)
vitor_resultado["Ranking"] = range(1, len(vitor_resultado) + 1)

print(vitor_resultado[["Ranking", "Categoria", "Região", "Quantidade", "Taxa_devolucao", "Produto mais devolvido"]])
```

```python
# tabela já filtrada para Vítor Rodrigues
vitor_resultado = pd.DataFrame({
    "Categoria": ["Material de escritório", "Móveis", "Tecnologia"],
    "Região": ["Norte", "Norte", "Norte"],
    "Quantidade": [2230, 980, 759],
    "Taxa_devolucao": [0.451570, 0.454082, 0.450593],
    "Produto mais devolvido": [
        "Acme Tesouras, Retangular",
        "Novimex Protetor de piso, Ajustável",
        "Hewlett Tinta, Vermelho"
    ]
})

# Defina parâmetros financeiros
preco_medio = {
    "Material de escritório": 100,   # exemplo: R$ 100 por pedido
    "Móveis": 500,                   # exemplo: R$ 500 por pedido
    "Tecnologia": 300                # exemplo: R$ 300 por pedido
}
percentual_desconto = 0.10  # 10% de desconto nas devoluções

# Calcula devolvidos e não devolvidos
vitor_resultado["Devolvidos"] = (vitor_resultado["Quantidade"] * vitor_resultado["Taxa_devolucao"]).round()
vitor_resultado["Nao_devolvidos"] = vitor_resultado["Quantidade"] - vitor_resultado["Devolvidos"]

# Calcula valores financeiros
vitor_resultado["Preco_medio"] = vitor_resultado["Categoria"].map(preco_medio)
vitor_resultado["Faturamento_bruto"] = vitor_resultado["Quantidade"] * vitor_resultado["Preco_medio"]
vitor_resultado["Valor_devolvido"] = vitor_resultado["Devolvidos"] * vitor_resultado["Preco_medio"]
vitor_resultado["Desconto"] = vitor_resultado["Valor_devolvido"] * percentual_desconto
vitor_resultado["Faturamento_liquido"] = vitor_resultado["Faturamento_bruto"] - vitor_resultado["Valor_devolvido"] - vitor_resultado["Desconto"]

print(vitor_resultado[[
    "Categoria", "Região", "Quantidade", "Devolvidos", "Nao_devolvidos",
    "Faturamento_bruto", "Valor_devolvido", "Desconto", "Faturamento_liquido"
]])
```

## Vítor Rodrigues

```python
# Filtra apenas Vítor Rodrigues
vitor_df = df_final[df_final["Gerente regional"].str.strip() == "Vitór Rodrigues"]

# Agrupa por categoria
vitor_resultado = vitor_df.groupby("Categoria").agg(
    Quantidade=("Quantidade", "sum"),
    Vendas=("Vendas", "sum"),
    Desconto_total=("Desconto", "sum"),
    Lucro_total=("Lucro", "sum"),
    Devolvidos=("Devolvido", "sum")
).reset_index()

# Calcula não devolvidos
vitor_resultado["Nao_devolvidos"] = vitor_resultado["Quantidade"] - vitor_resultado["Devolvidos"]

# Impacto financeiro das devoluções
vitor_resultado["Valor_devolvido"] = (
    vitor_resultado["Vendas"] * (vitor_resultado["Devolvidos"] / vitor_resultado["Quantidade"])
).round()

# Faturamento líquido (já considerando devoluções e descontos)
vitor_resultado["Faturamento_liquido"] = (
    vitor_resultado["Vendas"] - vitor_resultado["Valor_devolvido"] - vitor_resultado["Desconto_total"]
)
pd.set_option("display.float_format", "{:,.2f}".format)

print(vitor_resultado[[
    "Categoria", "Quantidade", "Vendas", "Devolvidos", "Nao_devolvidos",
    "Desconto_total", "Lucro_total", "Valor_devolvido", "Faturamento_liquido"
]])
```
